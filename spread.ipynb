```python
import pandas as pd
import numpy as np
import seaborn as sns

from IPython.display import display
import datetime

matches_table = pd.read_csv('data/matches.csv')
display(matches_table)
```

```python
set_players = sorted(set(matches_table['Player 1']).union(matches_table['Player 2']))
display(set_players)
```

```python
spread_table = pd.DataFrame(np.zeros([5, 5]))
spread_table.replace(0, np.NaN, inplace=True)
spread_table.index = set_players
spread_table.columns = set_players

for (i, row) in matches_table.iterrows():
    (p1, p2) = sorted([row['Player 1'], row['Player 2']])
    spread_table.loc[p1, p2] = row['Score 1'] - row['Score 2']

display(spread_table)
```

```python
ppg_table = pd.DataFrame()
ppg_table.index = set_players
ppg_table['Points For'] = 0
ppg_table['Points Against'] = 0
ppg_table['Point Difference'] = 0
ppg_table['Games Played'] = 0

for (i, row) in matches_table.iterrows():
    ppg_table.loc[row['Player 1'], 'Points For'] += row['Score 1']
    ppg_table.loc[row['Player 1'], 'Points Against'] += row['Score 2']
    ppg_table.loc[row['Player 2'], 'Points For'] += row['Score 2']
    ppg_table.loc[row['Player 2'], 'Points Against'] += row['Score 1']
    ppg_table.loc[row['Player 1'], 'Games Played'] += 1
    ppg_table.loc[row['Player 2'], 'Games Played'] += 1
    ppg_table['Point Difference'] = ppg_table['Points For'] - ppg_table['Points Against']

ppg_table['PPG'] = ppg_table['Point Difference'] / ppg_table['Games Played']

display(ppg_table)
```
